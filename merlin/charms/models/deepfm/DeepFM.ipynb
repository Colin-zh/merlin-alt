# Motivation
论文原文：[《DeepFM: A Factorization-Machine based Neural Network for CTR Prediction》](https://arxiv.org/abs/1703.04247)

参考学习地址：[速览 DeepFM: 使用 FM 取代 Wide & Deep 中的 LR](https://zhuanlan.zhihu.com/p/57158486)

诸如FNN、PNN等NN模型专注于隐式的高阶特征相关性，而LR、FM则专注于显式的低阶特征相关性。Wide&Deep同时考虑了两者，但Wide部分需要人工参与特征工程，所以 **<font color="red">DeepFM动机就是用FM取代Wide的LR</font>**。

 # 一些思考
 * FM与NN层共享Embedding，有效降低模型复杂度，同时在embedding的学习中同时接受来自“low & high order interaction”部分的反馈
 * DeepFM的思路与PNN思路很像，都引入了pairwise field的二阶相关性，差异在于PNN将得到的二阶特征直接喂给NN构建high-order，而DeepFM拿来作为final low-order特征参与最后计算
