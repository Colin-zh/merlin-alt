## Field
FFM引入了域（Field）的概念，可看作是对特征进行分组。e.g. 对于性别而言通常有female和male两种取值，对取值进行one-hot后，性别特征被拆分为两个独立的特征$x_{female}$和$x_{male}$。这两个“独立”特征本源于一个特征域（性别），若按照FM的思路则会有$x_{female, male}$这样必不会出现的交叉项。考虑到共域问题后的FFM公式定义如下：
$$
\begin{align}
y=w_0+\Sigma_{i=1}^nw_ix_i+\Sigma_{i=1}^{n-1}\Sigma_{j=i+1}^n<V_{i,f_j}, V_{j,f_i}>x_ix_j
\end{align}
$$

其中，$f$为域映射函数，$f_i$表示$x_i$特征对应的field编号。相较于FM，FFM的唯一差异就是在二阶交叉的隐向量中增加了域标识，假设Field类目数为$F$，那么<font color="red"> 对于每个特征 $x_i$ 拥有 $F$ 个对应的隐向量，分别应用于与不同域特征进行交叉</font>。设隐向量维度为$k$，FFM二阶交叉项参数为$nkF$。

求解时，由于引入了Field，不能再使用对称矩阵分解的trick，所以FFM模型推断时的时间复杂度为$O(kn^2)$，另隐向量表示为$V_{i,f_j}=(v_{i,f_j}^1,v_{i,f_j}^2,...,v_{i,f_j}^k)$，原公式展开有
$$
\begin{align}
y=w_0+\Sigma_{i=1}^nw_ix_i+\Sigma_{i=1}^{n-1}\Sigma_{j=i+1}^n\Sigma_{q=1}^kv_{i,f_j}^qv_{j,f_i}^qx_ix_j
\end{align}
$$
